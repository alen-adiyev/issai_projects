```python
#writing text by seperate segmented files

for j in range (1,65):
    with open('result.'+ str(j) + '.json') as f:
        data = json.load(f)
    data_list = list(data['utts'])
    for i in range(0, len(data['utts'])):
        text = data['utts'][data_list[i]]['output'][0]['rec_text']
        f = open(data_list[i][30:-4]+".txt", "w")
        f.write(text[:-5])
        f.close()
```

```python
#full transcription for the audio
f = open("interview_2_new.txt", "w")
for i in range(1, 512):
    text = data['utts']['/home/segmented_cs/interview_2_new/chunk_'+str(i)+'.wav']['output'][0]['rec_text']
    f.write("[" + str(i) + "] " + text[:-5] + "\n")
f.close()
```
